```python
import pandas as pd
df = pd.read_csv("44a688d7-8cd5-4c31-baa5-ec9fee468092.csv")
df
```

```python
# Gerar um novo dataframe somente os dados do Brazil, ordenados pela localização e ano (maior para menor)
df_ordenado = df[df['Location'] == 'Brazil'].sort_values(by=['Period'], ascending=False)
df_ordenado
```

```python
df_ordenado = df_ordenado.reset_index(drop=True)
df_ordenado
```

```python
df_ordenado['Local'] = ''
df_ordenado
```

```python
len(df_ordenado)
```

```python
tipos = ['Towns', 'Total', 'Urban', 'Cities', 'Rural']
for i in range(len(df_ordenado)):
    df_ordenado.at[i, 'Local'] = tipos[i % len(tipos)]
    #print(tipos[i % len(tipos)])
    #print(df_ordenado.at[i, 'FactValueNumericHigh'])
df_ordenado
```

```python
df_ordenado['valor_central'] = 0.0

for i in range(len(df_ordenado)):
    df_ordenado.at[i, 'valor_central'] = float(df_ordenado.at[i, 'Value'][:df_ordenado.at[i, 'Value'].find('[')])
df_ordenado

 
```

```python
#df_ordenado[df_ordenado['Period']==2019].Value
#pm2,5 total do Brasil por ano
import matplotlib.pyplot as plt

x = df_ordenado[df_ordenado['Local']=='Total']['Period']
y = df_ordenado[df_ordenado['Local']=='Total']['valor_central']
plt.bar(x,y)
plt.show()
```

```python
selected_columns = ['Location', 'ValueType', 'FactValueNumeric', 'valor_central']
df_selected = df[selected_columns]

# Filtrando apenas os dados relacionados ao índice de PM2.5
df_pm25 = df_selected[df_selected['ValueType'] == 'PM2.5']

# Agrupando por país e calcular a média dos valores de PM2.5
df_pm25_grouped = df_pm25.groupby('Location')['FactValueNumeric'].mean().reset_index()

# Ordenando os países pelos maiores valores de PM2.5
df_pm25_top3 = df_pm25_grouped.nlargest(3, 'FactValueNumeric')

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(df_pm25_top3['Location'], df_pm25_top3['FactValueNumeric'], color='skyblue')
plt.xlabel('Países')
plt.ylabel('Média de PM2.5')
plt.title('Top 3 Países com Maiores Índices de PM2.5')
plt.show()
```

```python
selected_columns = ['Location', 'ValueType', 'valor_central']
df_selected = df[selected_columns]

# Filtrando apenas os dados relacionados ao índice de PM2.5
df_pm25 = df_selected[df_selected['ValueType'] == 'PM2.5']

# Agrupando por país e calculando a média dos valores de PM2.5
df_pm25_grouped = df_pm25.groupby('Location')['valor_central'].mean().reset_index()

# Ordenando os países pelos menores valores de PM2.5
df_pm25_bottom3 = df_pm25_grouped.nsmallest(3, 'valor_central')

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(df_pm25_bottom3['Location'], df_pm25_bottom3['valor_central'], color='skyblue')
plt.xlabel('Países')
plt.ylabel('Média de PM2.5')
plt.title('Top 3 Países com Menores Índices de PM2.5')
plt.show()
```
