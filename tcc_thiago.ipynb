```python
#df_ordenado[df_ordenado['Period']==2019].Value
#pm2,5 total do Brasil por ano
import matplotlib.pyplot as plt

x = df[(df['Dim1']=='Total') & (df['Location']=='Brazil')]['Period']
y_total = df[(df['Dim1']=='Total') & (df['Location']=='Brazil')]['FactValueNumeric']
y_rural = df[(df['Dim1']=='Rural') & (df['Location']=='Brazil')]['FactValueNumeric']
y_cities = df[(df['Dim1']=='Cities') & (df['Location']=='Brazil')]['FactValueNumeric']
y_towns = df[(df['Dim1']=='Towns') & (df['Location']=='Brazil')]['FactValueNumeric']
y_urban = df[(df['Dim1']=='Urban') & (df['Location']=='Brazil')]['FactValueNumeric']
plt.bar(x,y_total, label="Total")
plt.bar(x,y_rural, label="Rural")
plt.bar(x,y_cities, label="Cities")
plt.bar(x,y_towns, label="Towns")
plt.bar(x,y_urban, label="Urban")
plt.legend()
plt.show()
```

```python
df.groupby('Dim1')['Dim1'].count()
```

```python
# Agrupando por país e calcular a média dos valores de PM2.5
x = df[df.Dim1 == 'Total'].groupby('Location')['FactValueNumeric'].mean().sort_values(ascending=False).head(3).index
y = df[df.Dim1 == 'Total'].groupby('Location')['FactValueNumeric'].mean().sort_values(ascending=False).head(3).values


# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(x,y, color='skyblue')
plt.xlabel('Países')
plt.ylabel('Média de PM2.5')
plt.title('Top 3 Países com Maiores Índices de PM2.5')
plt.show()

```

```python
# Agrupando por país e calcular a média dos valores de PM2.5
x = df[df.Dim1 == 'Total'].groupby('Location')['FactValueNumeric'].mean().sort_values(ascending=True).head(3).index
y = df[df.Dim1 == 'Total'].groupby('Location')['FactValueNumeric'].mean().sort_values(ascending=True).head(3).values


# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(x,y, color='skyblue')
plt.xlabel('Países')
plt.ylabel('Média de PM2.5')
plt.title('Top 3 Países com menores Índices de PM2.5')
plt.show()
```
